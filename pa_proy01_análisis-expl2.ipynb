```python
#par√°metros xgboost

params = {
    'objective' :'reg:linear',
    'max_depth' :1,
    'silent' :1,
    'eta' :0.5
}

num_rounds = 5
```

```python
watchlist = [(dtest,'test'),(dtrain,'train')]
bst = xgb.train(params, dtrain, num_rounds, watchlist)
```

```python
xgb.plot_importance(bst,importance_type='gain',xlabel='Gain')
```
